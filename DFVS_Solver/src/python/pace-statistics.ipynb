# Imports & Config

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Style plot
plt.style.use('default')
font = {'size' : 32}
plt.rc('font', **font)
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)
```

# Load data

```python
df_0 = pd.read_csv("../logs/PaceLog-0-ClassicReductions.csv", sep=",", header=0)
```

# Scatter plot

```python
fig, ax = plt.subplots(figsize=(25.6, 14.4))
ax.scatter(df_0.index, df_0['n'], c='tab:blue', marker='o', alpha=0.5, s=200, label='no reductions')
ax.scatter(df_0.index, df_0['reduce_n'], c='tab:orange', marker='o', alpha=0.5, s=200, label='classic reductions')

ax.set_xlabel('optimal k')
#ax.set_xscale('log')
#ax.set_yscale('log')
#ax.set_xlim([1, 70000000])
#ax.set_ylim([1, 70000000])
ax.set_xlabel('graph id', labelpad=20)
ax.set_ylabel('node count', labelpad=20)
#ax.tick_params(axis='both', length=20, which='major', pad=12)
ax.legend()
ax.grid()

fig.savefig('plots/v2/ScatterPlotForbiddenNodes.png',  bbox_inches='tight')
fig.show()
```
